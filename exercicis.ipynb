````python
from typing import Literal

import numpy as np

Fitxers = Literal["data/sistema.npz"] | Literal["data/sistemaedd.npz"] | Literal["data/sistemasp.npz"]

def importar_dades(fitxer: Fitxers):
    """
    Retorna la matriu A i el vector b contingut en el fitxer
    especificat.
    """
    data = np.load(fitxer)
    print(f"Fichero: {fitxer}\nClaves: {data.files}\n")
    A = data[data.files[0]]
    #print(f"Matriu A:\n{A}\n")
    b = data[data.files[1]]
    #print(f"Vector b:\n{b}\n")
    data.close()

    return A, b
````
