# Tasca 1. Mètodes directes

## Paràmetres

````python
IMPRIMIR_MATRIUS = False
DIRECTORI_DADES = "data/"
````

## Càrrega de dades

````python
from fitxers import Fitxers

import numpy as np


def importar_dades(fitxer: Fitxers):
    """
    Retorna la matriu A i el vector b contingut en el fitxer
    especificat.
    """
    data_ = np.load(DIRECTORI_DADES + fitxer)
    print(f"Fitxer: {fitxer}\nClaus: {data_.files}\n")
    A_ = data_[data_.files[0]]
    b_ = data_[data_.files[1]]
    if IMPRIMIR_MATRIUS:
        print(f"Matriu A:\n{A_}\n")
        print(f"Vector b:\n{b_}\n")
    data_.close()

    return A_, b_
````

Apliqueu la funció `lu` per donar la factorització $LU$ de la matriu $A$. A
partir de la matriu tornada per la funció `lu`, obtingueu explícitament les
matrius $L$ i $U$. Calculeu també la norma del vector residual de la
descomposició, 
$$
    \mathtt{res} := \| A - LU \|_{\infty}.
$$
Amb la factorització $LU$ trobada de la matriu $A$, calculeu, fent ús de la
funció `sol_lu`, la solució, $\mathring{x}$, del sistema $A x = b$. Doneu també
la norma del vector residual, 
$$
 \mathtt{res} := \left\| b - A \mathring{x} \right\|_{\infty}.
$$
